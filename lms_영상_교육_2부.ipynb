```python
m = folium.Map(location= [33.37329056367668, 126.53190545103804]) # 한라산 좌표

tooltip = '클릭해보세요!'

folium.Marker(
    [33.37329056367668, 126.53190545103804], # 한라산 좌표
    popup = '<strong>한라산</strong>',
    tooltip=tooltip
).add_to(m)

folium.Marker(
    [33.50779572779729, 126.49281515327071], # 제주 국제공항 좌표
    popup = '<strong>제주국제공항</strong>',
    icon = folium.Icon(color='#ff0000', icon = 'info-sign'),
    tooltip=tooltip
).add_to(m)

m
```

```python
import pandas as pd

# 공공데이터포털에서 '제주특별자치도 서귀포시_이용가능오름현황' 엑셀 사용

# csv 파일 호출 -> 오름 변수에 데이터시트 저장
오름 = pd.read_csv('서귀포시오름현황.csv', encoding = 'cp949')
오름.head()
```

```python
오름.iloc[0]
오름.iloc[0]['오름명']
오름.iloc[0]['위도']
오름.iloc[0]['경도']
len(오름)
```

```python
for i in range(len(오름)):
    print(오름.iloc[i]['오름명'], 오름.iloc[i]['위도'], 오름.iloc[i]['경도'])
```

```python
m = folium.Map(location= [33.37329056367668, 126.53190545103804])

tooltip = '클릭해보세요!'

for i in range(len(오름)):
    folium.Marker(
        [오름.iloc[i]['위도'], 오름.iloc[i]['경도']],
        popup = folium.Popup(f'<strong>{오름.iloc[i]["오름명"]}</strong>' , max_width = 500), # 마커 클릭 시 팝업, 길이 설정으로 가로로 출력
        tooltip=tooltip
    ).add_to(m)

m
```

```python
folium.Circle(
    [33.37329056367668, 126.53190545103804],
    radius = 10000,
    color = 'red',
    fill = True # 색상을 반투명으로 칠할 것인지
).add_to(m)

m
```

```python
# 직선거리 출력
folium.PolyLine(
    locations = [[33.37329056367668, 126.53190545103804], [33.3017594, 126.5573394]],
    tooltip = '한라산에서 다른 오름까지'
).add_to(m)

m
```
