# 크롤링


## URL(Uniform Resource Locator)
* 자원이 어디 있는지 알려주기 위한 규약
* 흔히 웹사이트 주소로 알고 있지만, URL은 웹 사이트 주소뿐만 아니라 컴퓨터 네트워크상의 자원을 모두 나타낼 수 있다.
* 그 주소에 접속하려면 해당 URL에 맞는 프로토콜을 알아야하고, 그와 동일한 프로토콜로 접속(FTP 프로토콜인 경우에는 FTP 클라이언트를 이용해야 하고, HTTP인 경우에는 웹 브라우저를 이용해야한다. 텔넷의 경우에는 텔넷 프로그램을 이용해서 접속)

## HTTP(Hypertext Transger Protocol)
* HTML, XML, Javascript, 오디오, 비디오, 이미지, PDF, Etc
* 요청 또는 상태 라인 / 해더(생략가능) / 빈줄(해더의 끝) / 바디(생략가능)

### HTTP 처리방식
* GET : 리소스 취득(? 뒤에 이어붙이는 방식 - 작은 값들)
* POST : 리소스 생성(Body에 붙이는 방식 - 상대적으로 큰 용량)
* PUT, DELETE, HEAD, OPTIONS, TRACE, CONNECT

### 상태코드
* 200
* 201
* 301
* 403
* 404
* 500
* 503


---


* 200번대 : 성공을 의미
* 400번대 : 클라이언트 오류, 요청 문법이 잘못되었거나 요청을 처리할 수 x
* 500번대 : 서버오류, 서버가 명백히 유효한 요청에 대해 충족을 실패했다

## Selector
* 태그에 좀 더 세밀한 접근이 가능
* class를 지칭할 때는 '.'을 사용하고, id를 지칭할 때는 '#'를 사용
* 탐색하고자 하는 태그가 특정태그 하위에 있을때 '>'를 사용

```python
import requests
from bs4 import BeautifulSoup

response = requests.get('http://www.paullab.co.kr/stock.html')

response.encoding = 'utf-8'
html = response.text

soup = BeautifulSoup(html, 'html.parser')
```

```python
soup.select('#update')
```

```python
soup.select('.table > tbody > tr') # table class안에 tbody안에 모든 tr 태그 출력
```

```python
soup.select("p > a:nth-of-type(2)") # p태그 안에 a태그안에 2번째 요소 호출
soup.select("p > a:nth-of-type(2)") # p태그 안에 a태그안에 짝수 요소 호출
soup.select("#link1 + .sister") # id와 class를 동시에 가진 것을 호출
```

```python
oneStep = soup.select('.main')[0]
oneStep
```
