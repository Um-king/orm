# 데이터프레임 연결

```python
import pandas as pd

# Sample DataFrames
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})
```

```python
# 가로로 연결 (열 방향)
# 주의할점 - df1과 df2가 동일한 행을 가지고 있을때만 가능함(동일한 행과 동일한 열의 개수여야한다.)

result = pd.concat([df1, df2], axis=1) # axis=1을 설정하여 열 방향으로 연결합니다.

print(result)
```

```python
# 열 이름이 다를 때 가로 연결

import pandas as pd

# Sample DataFrames with different column names
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'X': [5, 6], 'Y': [7, 8]})

# 가로로 연결 (열 방향) - 열 이름 다름
# 만약 데이터프레임의 열 이름이 다른 경우, ignore_index=True 옵션을 사용하여 열 이름을 재설정할 수 있습니다.
result = pd.concat([df1, df2], axis=1, ignore_index=True)

print(result)
```

```python
import pandas as pd

# Sample DataFrames
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})
```

```python
# 세로로 연결 (행 방향)
result = pd.concat([df1, df2]) # 기본적으로 axis=0이 설정되어있고 행을 기준으로 연결
# axis=0이 기본값이고 빠져있는 상태

print(result)
```

```python
# 인덱스가 중복되지 않도록 설정하려면 ignore_index=True 옵션을 사용합니다.
# 이때 사용하는것이 아주 유용 -> 행의 개수를 확인할 수도 있음

# 어떨때 사용할까요?
# 2023년 1월 ~ 2024년 1월 -> 분석을 열심히 했다.
# 신규로 2024년 2월을 등록할 때

import pandas as pd

# Sample DataFrames with different index
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]}, index=[0, 1]) # index 설정으로 어느 행에 할당할지?
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]}, index=[2, 3]) # 2행과 3행에 해당 값을 생성?

# 세로로 연결 (행 방향) - 인덱스 다름
result = pd.concat([df1, df2], ignore_index=True)

print(result)

```

```python
import pandas as pd

# 문자형을 숫자로 변경하는 방법 1. astype, 2. to_numberic
df['Stars'] = pd.to_numeric(df['Stars'], errors='coerce') # errors = 'coerce' 에러 처리방법 설정
```

```python
df.info() # Stars의 데이터 타입 변경 확인
```

```python
#IQR 기반 이상치 제거
Q1 = df['Stars'].quantile(0.25)
Q3 = df['Stars'].quantile(0.75)
IQR = Q3 - Q1
df = df[~((df['Stars'] < (Q1 - 1.5 * IQR)) | (df['Stars'] > (Q3 + 1.5 * IQR)))]
```

```python
# 박스플롯을 사용한 이상치 탐지
import matplotlib.pyplot as plt

dataset = pd.read_csv('/content/The-Big-List-20231127-Reviews-to-4650.csv', encoding='latin1')  # 인코딩이 필요한 경우 변경

# 'Stars' 컬럼의 비숫자 값을 NaN으로 변환
dataset['Stars'] = pd.to_numeric(dataset['Stars'], errors='coerce')

# NaN 값 제거
dataset = dataset.dropna(subset=['Stars'])

# 'Stars' 컬럼에 대한 박스플롯 생성
plt.boxplot(dataset['Stars'])
plt.title('Boxplot of Stars Ratings')
plt.ylim(1, 5)  # y축 범위를 1에서 5로 설정
plt.show()
```
