```python
# 31. 다음 리스트의 내장함수의 시간 복잡도가 O(1)이 아닌것은?

'''
1)  l[i]
2)  l.append(5)
3)  l[a:b]
4)  l.pop()
5)  l.clear()
'''

'''
>> 결과
3번 / 슬라이싱 되는 요소들 수 만큼 비례

1)  l[i]          => 시간복잡도 0(1)
2)  l.append(5)   => 시간복잡도 0(1)
3)  l[a:b]        => 슬라이싱 되는 요소들 수 만큼 비례
4)  l.pop()       => 시간복잡도 0(1)
5)  l.clear()     => 시간복잡도 0(1)
'''

'''
시간복잡도 = 문제를 해결하는데 걸리는 시간과 입력의 함수 관계
즉 한마디로 시간이 얼마나 걸리느냐..!

파이썬의 기본 산술 연산들은 O(1)으로 즉각적인 결과가 나타남
참고 사이트: https://chancoding.tistory.com/43
'''
```
